# Training the model using tweets dataset and using multinomial Naive Bayes model for sentiment analysis

```python
import pandas as pd
data=pd.read_csv("C:\\Users\\Administrator\\Downloads\\text_emotion.csv" ,encoding='latin-1',names=['emotions','value', 'tweets' ],skiprows=1)

data.drop('value',axis=1,inplace=True)
data.drop(0,axis=0)


data.emotions=data.emotions.str.replace('happiness','1')
data.emotions=data.emotions.str.replace('enthusiasm','1')
data.emotions=data.emotions.str.replace('relief','1')
data.emotions=data.emotions.str.replace('sadness','0')
data.emotions=data.emotions.str.replace('boredom','0')
data.emotions=data.emotions.str.replace('neutral','2')
data.emotions=data.emotions.str.replace('love','1')
data.emotions=data.emotions.str.replace('worry','0')
data.emotions=data.emotions.str.replace('fun','1')
data.emotions=data.emotions.str.replace('empty','2')
data.emotions=data.emotions.str.replace('hate','0')
data.emotions=data.emotions.str.replace('anger','0')
data.emotions=data.emotions.str.replace('surprise','3')
data.emotions.value_counts()
data['emotions']=data.emotions.astype(int)

data.head()
```

# Using Count vectorization to convert text to vector form(sparse matrix) then predicting the sentiment of the Donald Trump related tweets.
## Total tweets Trump related tweets are 913 and value of "c" tells the total positive tweets
